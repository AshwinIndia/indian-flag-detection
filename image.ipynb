# crop all images to make them square and resize them to 128x128

```python
import os
import cv2
import numpy
import sys
import time

def crop_and_resize(image_path, output_path):
    image = cv2.imread(image_path)
    height, width, channels = image.shape
    if height > width:
        diff = height - width
        image = image[diff//2:height-diff//2, 0:width]
    elif width > height:
        diff = width - height
        image = image[0:height, diff//2:width-diff//2]
    image = cv2.resize(image, (128, 128))
    cv2.imwrite(output_path, image)

def rename_images(folder_path,output_folder):
    for i, filename in enumerate(os.listdir(folder_path), start=1):
        if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
            image_path = os.path.join(folder_path, filename)
            output_path = os.path.join(output_folder, f"{i}.jpg")
            crop_and_resize(image_path, output_path)
            print(f"Resized {filename} to {i}.jpg")
```

```python
folder_path=r"similar_to_india"


output_folder=r"similar_to_india_resized"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for filename in os.listdir(folder_path):
    print("In For", filename)
    if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
        print("Processing", filename)
        image_path = os.path.join(folder_path, filename)
        output_path = os.path.join(output_folder, filename)
        crop_and_resize(image_path, output_path)
        print("Resized", filename)
```

```python
rename_images(folder_path,output_folder)
```
